```python
# importando bibliotecas

import pandas as pd
import numpy as np
```

```python
# importando dataset

fifa19 = pd.read_csv(r'fifa19.csv')
fifa19
```

```python
# criando a tabela dinâmica com a metrica média

# sintaxe: pd.pivot_table(dataset, index=, aggfunc=)

fifa19_filtrado = fifa19[fifa19['Nationality'].isin(['Argentina', 'Brazil', 'France', 'Germany', 'Italy'])]
pd.pivot_table(fifa19_filtrado, index='Nationality', aggfunc = np.mean)
```

```python
# criando a tabela dinamica com a metrica mediana

pd.pivot_table(fifa19_filtrado, index='Nationality', aggfunc= np.median)
```

```python
# criando uma nova pivot table com as colunas: 'Age', 'Overall', 'Acceleration' e 
#o index sendo a coluna 'Nationality'

pd.pivot_table(fifa19_filtrado, index='Nationality',
              values=['Age', 'Overall', 'Acceleration'],
              aggfunc = np.median)
```

```python
# criando uma nova pivot table com dois index 'Nationality', 'Club' e gerando um 
# arquivo .CSV

pivot_table = pd.pivot_table(fifa19_filtrado, index=['Nationality','Club'],
              values=['Age', 'Overall', 'Acceleration'],
              aggfunc = np.median)

pivot_table.to_csv('pivot_table.csv', sep=';', decimal=',')

# o arquivo estara dentro da pasta onde está localizado o .ipynb
```

```python
# um outro exemplo de pivot table com o campo columns ='Position'

pd.pivot_table(fifa19_filtrado, index=['Nationality', 'Club'],
              values=['Age', 'Overall', 'Acceleration'],
              columns='Position',
              aggfunc = np.median)
```

```python
# um outro exemplo de pivot table com o campo columns ='Club'

pd.pivot_table(fifa19_filtrado, index=['Nationality'],
              values=['Age', 'Overall', 'Acceleration'],
              columns='Club',
              aggfunc= np.median)
```
