```python
# importando as bibliotecas
import seaborn as sns
import pandas as pd
```

```python
# carregando arquivo
iris = pd.read_csv(r"iris_csv.csv")
iris.head()
```

## Gráficos de Dispersão

```python
# Relational plot
sns.relplot(x="sepallength", y="sepalwidth", data=iris);
```

```python
# muda cor de acordo com grupo (class)
sns.relplot(x="sepallength", y="sepalwidth", hue='class' ,data=iris);
```

```python
# muda estilo do ponto de acordo com grupo (especie), ajuda a enfatizar
sns.relplot(x="sepallength", y="sepalwidth", hue='class' , style='class',data=iris);
```

```python
# muda cor de acordo com numero
sns.relplot(x="sepallength", y="sepalwidth", hue='petallength',data=iris);
```

```python
# muda tamanho do ponto de acordo com tamanho da petala
sns.relplot(x="sepallength", y="sepalwidth", size='petallength',data=iris);
```
