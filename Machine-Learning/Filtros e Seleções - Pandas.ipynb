```python
# importando biblioteca

import pandas as pd
```

```python
# lendo o dataframe fifa 19

fifa19 = pd.read_csv(r'fifa19.csv')
fifa19.head()
```

```python
# usando o comando loc, passamos apenas os indices das linhas

fifa19.loc[0:5]
```

```python
# seleção com os indices das linhas e nome da coluna

fifa19.loc[0:10, 'Name']
```

```python
# usando o comando iloc
# realizamos o filtro com o numero das linhas e colunas

filtro_iloc = fifa19.iloc[0:10,0:3]
filtro_iloc
```

```python
# filtrando as 5 primeiras linhas com todas as colunas

fifa19.iloc[0:5, :]
```

```python
# filtrando as colunas Age e Overall
# as linhas onde o Age é maior que 30 ou Overall maior que 90

fifa19.loc[(fifa19['Age']>30) | (fifa19['Overall']>90)]

# e: &
# ou: |
```

```python
# usando o comando query
# filtrando as linhas onde Age é maior que 30

fifa19.query('Age > 30')
```

```python
# filtrando as linhas que Age é maior que 30 E Overall maior que 90

fifa19.query('Age > 30 & Overall > 90')
```

```python
# criando a mediana da coluna Overall

overall_mediana = fifa19['Overall'].median()
overall_mediana
```

```python
# filtrando as linhas da coluna overall que são maiores que a mediana

fifa19.query('Overall > @overall_mediana')
```

```python
# filtrando os jogadores onde a nacionalidade é Brasil

fifa19.query(" Nationality == 'Brazil' ")
```

```python
# filtrando as linhas onde na coluna Age é diferente de 30

fifa19[fifa19['Age'] != 30]
```

```python
# criando um novo dataframe com as colunas: 
#[ID','Age', 'Name', 'Overall', 'Nationality', 'Club']

fifa19_v2 = fifa19.filter(['ID', 'Age', 'Name', 'Overall', 'Nationality', 'Club'])
fifa19_v2.head()
```

```python
# filtrando as linhas onde na coluna Nationality estão os países listados:
# ['Brazil','Chile', 'Colombia', 'Peru']

fifa19[fifa19['Nationality'].isin(['Brazil','Chile', 'Colombia', 'Peru'])]
```
